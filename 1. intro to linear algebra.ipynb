```python
'''
A robber travels at 150Km/h and has a headstart of 5-mins. A cop travels at 180Km/h.
Question: At what distance the cop catches the robber?
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000) # start, finish , n points
```

```python
# distance travelled by robber
d_r = 2.5*t

# distance travelled by sheriff
d_s = 3*(t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in kilometers)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,d_r,c='green')
ax.plot(t,d_s,c='brown')
plt.axvline(x=30,color='purple',linestyle='--')
plt.axhline(y=75,color='purple',linestyle='--')
```

```python
'''
Example 2
Jill designs solar panel as a hobby. On April 1st Jill's Mark 1 begins generating power 1KJ/day.
On May 1st her Mark 2 design beigns generating 4KJ of power per day.
Question: what day is it when mark 2 design has generated as much total energy as the mark 1 design 
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.linspace(0,50,1000) # start, finish, n points 
```

```python
# days taken by mark 1
t_m1 = d
# days taken by mark2
t_m2 = 4*(d-30)
```

```python
fig ,ax = plt.subplots()
plt.title("Energy produced by 2 panels")
plt.xlabel("days")
plt.ylabel("Energy in joules")
ax.set_xlim([0,50])
ax.set_ylim([0,50])
ax.plot(d,t_m1,color='green')
ax.plot(d,t_m2,color='red')
plt.axvline(x=40,color='blue',linestyle='--')
plt.axhline(y=40,color='blue',linestyle='--')
```
